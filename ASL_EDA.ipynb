# ASL FOR KIDS: EDA

```python
import os
import cv2
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import misc
```

```python
letters = os.listdir("asl-alphabet/asl_alphabet_train")
# removing DS Store
letters = letters[1:]
```

```python
print("there are " + str(len(letters)) + " classes of images in our training set, with 3000 images in each class.")
print("The classes are as follows: " + str([i for i in letters]))
```

```python
a1 = cv2.imread("asl-alphabet/asl_alphabet_train/A/A1.jpg")
plt.imshow(a1)
plt.show();
print("A1's Shape = " + str(a1.shape) + ", 200x200 pixels")

r1 = cv2.imread("asl-alphabet/asl_alphabet_train/R/R1.jpg")
plt.imshow(r1)
plt.show();
print("R1's Shape = " + str(r1.shape) + ", 200x200 pixels")
```

```python
A_images = os.listdir("asl-alphabet/asl_alphabet_train/A")

avg_A_letter = [[0 for j in range(200)] for i in range(200)]
for i in A_images:
    letter = cv2.imread("asl-alphabet/asl_alphabet_train/A/" + str(i), 0)
    avg_A_letter = avg_A_letter + letter

avg_A_letter = avg_A_letter / len(A_images)

plt.matshow(avg_A_letter)
plt.title("Average A Image")
plt.show();

R_images = os.listdir("asl-alphabet/asl_alphabet_train/R")
avg_R_letter = [[0 for j in range(200)] for i in range(200)]
for i in R_images:
    letter = cv2.imread("asl-alphabet/asl_alphabet_train/R/"+ str(i), 0)
    avg_R_letter = avg_R_letter + letter

avg_R_letter = avg_R_letter / len(R_images)

plt.matshow(avg_R_letter)
plt.title("Average R Image")
plt.show();
```

As you can see, the two plots of the average image look pretty similar, even though the signs look very different. Now I am going to create average plots for all of the letters. 

### Matshow( ) of average image of each image class

```python
for l in letters: 
    images = os.listdir("asl-alphabet/asl_alphabet_train/" + str(l))
    avg_letter = [[0 for j in range(200)] for i in range(200)]
    for i in images:
        letter = cv2.imread("asl-alphabet/asl_alphabet_train/" + str(l) + "/" + str(i), 0)
        avg_letter = avg_letter + letter

    avg_letter = avg_letter / len(images)

    plt.matshow(avg_letter)
    plt.title("Average " + str(l) + " Image")
    plt.show();
```

There is a darker cloud that moves around the square, this is the hand that is making the sign. Also, our training images seem to have a similar background in all of them, this might not be the case for our real-world webcam data. 

For our model, we should look at using a bounding box to detect the hand, and then use our classification model to classify the sign from there. 

### Histograms of each image class

```python
for l in letters: 
    images = os.listdir("asl-alphabet/asl_alphabet_train/" + str(l))
    avg_letter = [[0 for j in range(200)] for i in range(200)]
    for i in images:
        letter = cv2.imread("asl-alphabet/asl_alphabet_train/" + str(l) + "/" + str(i), 0)
        avg_letter = avg_letter + letter

    avg_letter = avg_letter / len(images)
    plt.hist(avg_letter.ravel(),256,[0,256])
    plt.title("Average " + str(l) + " Image")
    plt.show();
```
