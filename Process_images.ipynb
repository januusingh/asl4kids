```python
img = load_img('asl-alphabet/asl_alphabet_test/A_test.jpg')
def read_image(file_path):
    # For image visualization
    im = np.array(Image.open(file_path))
    img = cv2.imread(file_path, cv2.IMREAD_COLOR) #cv2.IMREAD_GRAYSCALE
    return img

def plot_pic(img):
    # Plot openCV pic
    pic = read_image(img)    
    plt.figure(figsize=(5,5))
    plt.imshow(pic)
    plt.show()
```

```python
mypath='asl-alphabet/asl_alphabet_train'
train_imgs = []
for path, subdirs, files in os.walk(mypath):
    for file in files:
        img = read_image(os.path.join(path,file)).flatten()
        train_imgs.append(img)
print(len(train_imgs))
```
