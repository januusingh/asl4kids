```python
n = 15
data = orig_data.sort_values(by=['person','timestamp_event'])

counts = data.groupby("person").count().iloc[:,1]

counts = counts[counts > n]
counts
data = data[data["person"].isin(counts.index)]

data = data.groupby('person').apply(lambda x: x[:len(x)//n*n])
data = data.reset_index(drop = True)
```

```python
features = ['x','y','duration','change of page','change_in_x','change_in_y','distance','x_velocity','y_velocity']
```

```python

data = data[features + ['MAX-adhesion_convolve']]
newData = []
labels1 = []
labels2=[]
for i in range(data.shape[0]//n):
    temp = []
    for j in range(n):
        temp.extend(data.iloc[i*n+j,:-1])
    newData.append(temp)
    labels1.append(max(data.iloc[n*i:n*i+n,-1]))
    labels2.append(len(set(data.iloc[n*i:n*i+n,-1])) == 2)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(newData, labels2, test_size=0.2, random_state=1)
from sklearn import linear_model

# Name our logistic regression object
LogisticRegressionModel = linear_model.LogisticRegression()

# we create an instance of logistic Regression Classifier and fit the data.
print ('Training a logistic Regression Model..')
LogisticRegressionModel.fit(x_train, y_train)

training_accuracy=LogisticRegressionModel.score(x_train,y_train)
print ('Training Accuracy:',training_accuracy)

test_accuracy=LogisticRegressionModel.score(x_test,y_test)
print('Accuracy of the model on unseen test data: ',test_accuracy)
```
